# 导入两个库

```python
import numpy as np 
import matplotlib.pyplot as plt
```

# 生成系数矩阵A和b

## 生成系数矩阵A

```python
n=100    #定义阶数  PS：以下各算法不再修改n的值，均在此基础上进行。只需修改n的值再重新运行算法即可计算不同阶数的结果。

middle_array = np.empty(n, dtype=float)    #创建对角元数组
middle_array.fill(6)

upper_array = np.empty(n-1, dtype=float)   #创建上对角元数组
upper_array.fill(1)

down_array = np.empty(n-1, dtype=float)    #创建下对角元数组
down_array.fill(8)

A=np.diag(middle_array)+np.diag(upper_array,1)+np.diag(down_array,-1)  #生成矩阵A

```

## 生成矩阵b

```python
b=np.full(
    shape=n,
    fill_value=15,
    dtype=float)
b[0]=7
b[n-1]=14
```

## 显示矩阵A, b规模和矩阵A, b

```python
A.shape,b.shape
```

```python
A,b
```

# 进行顺序Gauss消元法

```python
row = A.shape[0]#记录A矩阵行数
col = A.shape[1]#记录A矩阵列数
row,col
```

## 消去过程

```python
for i in range(0, row):                         #从0到99行循环
        for j in range(i+1, col):               #将A矩阵第i行以下各行依次消去
            temp = A[j][i]/A[i][i]
            b[j]=b[j]-b[i] * temp               #将b矩阵各元素依此消去
            for k in range(0, col):
                A[j][k]=A[j][k]-A[i][k] * temp
                       
```

```python
A,b        #显示徐消去后的矩阵A和b
```

## 回代过程

```python
x=np.empty(n, dtype=float)    #定义x空数组
x[row-1]=b[row-1]/A[row-1][row-1]    #从x[n-1]到x[0]依次回代求解
for i in range(row-2,-1,-1):   
    x[i]=b[i]/A[i][i]
    for j in range(i+1,row):
        x[i]=x[i]-A[i][j]*x[j]/A[i][i]
```

```python
x        #显示线性方程组的解
```

# 进行列主元Gauss消元法

## 重新生成矩阵A和b（矩阵A和b在上一问时已发生了改变）

```python
A=np.diag(middle_array)+np.diag(upper_array,1)+np.diag(down_array,-1)
b=np.full(
    shape=n,
    fill_value=15,
    dtype=float)
b[0]=7
b[n-1]=14
```

## 显示矩阵A, b规模和矩阵A, b

```python
A.shape,b.shape
```

```python
A,b
```

# 进行列主元Gauss消元法

```python
row = A.shape[0]#记录A矩阵行数
col = A.shape[1]#记录A矩阵列数
row,col
```

## 消去过程

```python
for i in range(0, row):                         #行循环
        for j in range(i+1, col):
            if abs(A[j][i]) > abs(A[i][i]):
                t = np.copy(A[i])
                A[i] = A[j]
                A[j] = t      
                t = b[i]
                b[i] = b[j]
                b[j] = t

        for j in range(i+1, col):               #将A矩阵第i列的下三角部分消为0
            temp = A[j][i]/A[i][i]
            b[j]=b[j]-b[i] * temp               #将b矩阵各元素依此消去
            for k in range(0, col):
                A[j][k]=A[j][k]-A[i][k] * temp    

                
```

```python
A,b        #显示徐消去后的矩阵A和b
```

## 回代过程

```python
x=np.empty(n, dtype=float)       #定义x空数组
x[row-1]=b[row-1]/A[row-1][row-1]
for i in range(row-2,-1,-1):       #从x[n-1]到x[0]依次回代求解
    x[i]=b[i]/A[i][i]
    for j in range(i+1,row):
        x[i]=x[i]-A[i][j]*x[j]/A[i][i]
```

```python
x        #显示线性方程组的解
```

# 进行追赶法

## 重新生成矩阵A和b（矩阵A和b在上一问时已发生了改变）

```python
A=np.diag(middle_array)+np.diag(upper_array,1)+np.diag(down_array,-1)
b=np.full(
    shape=n,
    fill_value=15,
    dtype=float)
b[0]=7
b[n-1]=14
```

## 显示矩阵A, b规模和矩阵A, b

```python
A.shape,b.shape
```

```python
A,b
```

## 进行追赶法

```python
a1 = np.insert(down_array,0,0)  #创建a数组，第一位补0
b1 = middle_array               #创建b数组，由于已使用b数组，这里用b1替代，a1、c1同理
c1 = np.insert(upper_array,n-1,0)         #创建c数组,最后一位补0
```

```python
a1,b1,c1   #显示三个数组
```

```python
y = np.zeros(n)
B = np.zeros(n)
x = np.zeros(n)
```

```python
y[0] = b[0] / b1[0]
d = b1[0]    #α[0]
```

### 追

```python
for i in range(1, n):
    B[i - 1] = c1[i - 1] / d
    d = b1[i] - a1[i] * B[i - 1]   #α不再另创数组，每次循环修改值
    y[i] = (b[i] - a1[i] * y[i - 1]) / d
x[n-1] = y[n-1]
```

### 赶

```python
for i in range(n-2, -1, -1):
    x[i] = y[i] - B[i] * x[i + 1]
```

```python
x        #显示线性方程组的解
```

# Jacobi迭代法

## 显示矩阵A, b规模和矩阵A, b。
## 此时不需重新生成矩阵，矩阵A,b在追赶法中未改变。

```python
A.shape,b.shape
```

```python
A,b
```

## 进行Jacobi迭代

```python
x=np.zeros(shape=(n,100))  #创建x空矩阵，第几列代表第几次迭代的值，初值都取为0
x
```

```python
# 以分量形式进行Jacobi迭代
for k in range(1, 100):  #循环迭代，迭代99次
    for i in range(0, n):
        x[i][k]=b[i]/A[i][i]
        for j in range(0, i):
            x[i][k]=x[i][k]-A[i][j] * x[j][k-1]/A[i][i]
        for j in range(i+1, n):
            x[i][k]=x[i][k]-A[i][j] * x[j][k-1]/A[i][i]
```

```python
x   #显示x矩阵
```

```python
x[:,99] #显示迭代99次后的结果
```

## 进行Gauss-Seidel迭代

```python
x=np.zeros(shape=(n,100)) #创建x空矩阵，第几列代表第几次迭代的值，初值都取为0
x
```

```python
#以分量形式进行Jacobi迭代
for k in range(1, 100):   #循环迭代，迭代99次
    for i in range(0, n): 
        x[i][k]=b[i]/A[i][i]
        for j in range(0, i):
            x[i][k]=x[i][k]-A[i][j] * x[j][k]/A[i][i]
        for j in range(i+1, n):
            x[i][k]=x[i][k]-A[i][j] * x[j][k-1]/A[i][i]
```
